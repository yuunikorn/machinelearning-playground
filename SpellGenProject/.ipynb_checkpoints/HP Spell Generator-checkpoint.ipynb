Generating Fandom Names
================

This is a neural network for generating names. 

It uses the Python library [conx](https://conx.readthedocs.io) for specifying neural networks.

```python
import conx as cx
```

Reading in the data
-------------------

```python
def read_file(f):
    """Read names from a file into a list of strings/names. 
    The has one name per line."""
    names = []
    for line in f:
        line = line.strip()
        if not line.startswith("#") and line != "":
            names.append(line)
    return names

with open("Data/HPSpell.txt") as f:
    names = read_file(f)
    
print(names[:10])
```

```python
#checks if names are valid

for n in names:
    if (n.isalnum() == 0):
        print(n)
        
print("done")
```

Encoding and decoding strings
-----------------------------

```python
LETTERS = ord('z') - ord('a')
NONALPHA = LETTERS + 1
START = LETTERS + 2
END = LETTERS + 3
VOCAB_LEN = END + 1 # 'a' is encoded as 0
         
def encode(s):
    """Encode a single character into an integer or a string into 
    a list of of integers. """
    if len(s) == 1:
        if s.isalpha():
            return ord(s.lower()) - ord('a')
        else:
            return NONALPHA
    else:
        return [START] + cleanup([encode(c) for c in s]) + [END]

def cleanup(items):
    """Replace multiple non-alphabetic characters in a row with just one."""
    retval = []
    for i in items:
        if ((i != NONALPHA) or  
            (len(retval) == 0) or 
            (retval[-1] != NONALPHA)):
            retval.append(i)
    return retval

def decode(code):
    """Convert an integer into a character or a list of integers
    into a string."""
    if isinstance(code, (list, tuple)):
        return "".join([decode(c) for c in code if c != START and c != END])
    elif code == NONALPHA:
        return '-'
    else:
        return chr(ord('a') + code)

print(encode("Anna-Banana"))
print(decode(encode("An-na")))
```

Encode all names.

```python
encoded_names = [encode(n) for n in names]

for i in range(10):
    print(names[i], encoded_names[i])
```

Define the network
------------------

ConX allows you specify a network by just listing the layers. It comes with a lot of predefined layers. `cx.Layer` is a default feed-forward layer and `cx.EmbeddingLayer` and `cx.SimpleRNNLayer` are what their names suggest. (See the [documentation of conx.layers](https://conx.readthedocs.io/en/latest/modules.html) for more information.)

For each layer, you have to give a name and then you have to specify the dimensions of the layer. E.g. below the layer called "rnn" has 128 neurons.

Note that for training we are going to chop our names into stretches of <code>training_sequence_length</code> letters.

```python
def name_generation_network(network_name, training_sequence_length):
    net = cx.Network(network_name)
    net.add(cx.Layer("input", training_sequence_length),
            cx.EmbeddingLayer("embed", VOCAB_LEN, 130),
            cx.SimpleRNNLayer("rnn", 128),
            cx.Layer("output", VOCAB_LEN, activation="softmax"))
    net.connect()
    net.compile(loss='categorical_crossentropy', optimizer='adam')
    return net
```

Prepare the training data
-------------------------

```python
training_sequence_length = 20

def pad(vector, padding_length):
    """Pad the given name so that it has a total of padding_length 
    occurrences of START in the beginning and of END in the END."""
    return [START] * (padding_length-1) + name + [END] * (padding_length-1)
           
training_data = []
for name in encoded_names:
    name = pad(name, training_sequence_length)
    for i in range(len(name) - training_sequence_length):
        given = name[i:i+training_sequence_length]
        predicted = name[i+training_sequence_length]
        if len(given) == training_sequence_length:
            training_data.append([given, cx.onehot(predicted, VOCAB_LEN)])

for training_item in training_data[:3]:
    print(training_item)
```

Create a network and load in the training data
----------------------------------------------

```python
net1 = name_generation_network("spells", 20)
net1.dataset.load(training_data)
net1.summary()
```

Train the network
-----------------

```python
# Uncomment the following line and comment out the line after it, 
# if you want to read in a saved network.
#net1.load()
net1.train(80, save=True)
```

Generate names
--------------

```python
def generate(net):
    inputs = [START] * training_sequence_length
    generated_name = ""
    next_letter = 0
    while next_letter != END:
        outputs = net.propagate(inputs)
        next_letter = cx.choice(p=outputs)
        if next_letter!=END:
            generated_name += decode(next_letter)
        inputs = inputs[1:] + [next_letter]
    return generated_name
```
